# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
```

```python
# read input victoria images and show them in a row together
IMAGE_CNT = 7
FIG_SIZE = 16

fig = plt.figure(figsize=(FIG_SIZE, FIG_SIZE))

for i in range(1, IMAGE_CNT + 1):
    victoria_i = cv2.imread(f"images/victoria{i}.png")
    fig.add_subplot(1, IMAGE_CNT, i)
    plt.imshow(victoria_i)
    plt.axis('off')
    plt.title(f"Image {i}")
```

```python
# initialize OpenCV's image sticher object and then perform the image stitching on input images
 
image_sticher = cv2.Stitcher.create()
status, output = image_sticher.stitch(tuple(cv2.imread(f"images/victoria{i}.png") for i in range(1, IMAGE_CNT + 1)))
```

```python
# show victoria panorama
plt.imshow(output)
plt.axis('off')
plt.title(f"victoria panorama")
plt.show()
```
